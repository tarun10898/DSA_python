```python
class MyClass:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        print(f"MyClass instance created with name: {self.name} and age: {self.age}")

    def display_info(self):
        print(f"Name: {self.name}, Age: {self.age}")

# Create an instance of MyClass and pass values to its constructor
instance1 = MyClass("Alice", 30)

# Access attributes and call methods of the instance
print(f"Accessing name attribute: {instance1.name}")
instance1.display_info()

# Create another instance with different values
instance2 = MyClass("Bob", 25)
instance2.display_info()
```
