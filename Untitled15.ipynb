# Task
Identify unique names from the `df_unique` DataFrame, ensuring that for each name, the entry with the most available details (i.e., fewest missing values) is selected, and then display the first 10 such unique name entries along with a summary of the approach.

## Calculate Missing Values

### Subtask:
Calculate the number of missing values for each row in the `df_unique` DataFrame and store it in a new column called 'missing_count'.


**Reasoning**:
To calculate the number of missing values for each row, I will apply the `.isnull()` method to the `df_unique` DataFrame and then sum the resulting boolean values across each row using `sum(axis=1)`. The result will be stored in a new column named 'missing_count'.



```python
df_unique['missing_count'] = df_unique.isnull().sum(axis=1)
display(df_unique.head())
```

**Reasoning**:
The previous code produced a `SettingWithCopyWarning` because it was attempting to modify a view or copy of a DataFrame. To ensure the 'missing_count' column is added correctly to the original `df_unique` DataFrame and avoid the warning, I will explicitly use `.loc` for assignment.



```python
df_unique = df_unique.copy()
df_unique.loc[:, 'missing_count'] = df_unique.isnull().sum(axis=1)
display(df_unique.head())
```

## Select Rows with Most Details

### Subtask:
Group the DataFrame by the 'Name' column and for each group, select the row that has the minimum 'missing_count'. This ensures that for duplicate names, the entry with the most available details is retained.


**Reasoning**:
To select the row with the minimum 'missing_count' for each unique name, I will group the DataFrame by 'Name', find the index of the minimum 'missing_count' for each group using `idxmin()`, and then use these indices to select the corresponding rows from the `df_unique` DataFrame.



```python
idx_most_details = df_unique.groupby('Name')['missing_count'].idxmin()
df_most_details = df_unique.loc[idx_most_details]
display(df_most_details.head())
```
